```python
import numpy as np
import matplotlib.pyplot as plt
from lab_utils_common import plot_data, sigmoid, draw_vthresh
```

```python
#The dataset
X = np.array([[0.5, 1.5], [1, 1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
Y = np.array(([0, 0, 0, 1, 1, 1])).reshape(-1, 1)
```

```python
#Plotting

fig, ax = plt.subplots(1,1, figsize=(4,4))
plot_data(X, Y, ax)

ax.axis([0, 4, 0, 3.5])
ax.set_ylabel('$x_1$')
ax.set_xlabel('$x_0$')
plt.show()
```

```python
#Let's plot the sigmoig(z) function over a range of values from -10 to 10
z = np.arange(-10,11)

fig, ax = plt.subplots(1,1, figsize=(5,3))
#Plot z vs sigmoid(z)
ax.plot(z, sigmoid(z), c='r')

ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
#draw_vthresh(ax,0)
```

```python
#Plotting the equation -3 + x_0 + x_1 = 0 (this is a particular example, where: w_0 = 1, w_1 = 1, b = -3)

x0 = np.arange(0, 6)
x1 = 3 - x0

fig, ax = plt.subplots(1,1, figsize=(5,4))
#The decision boundary
ax.plot(x0,x1, c='g')
ax.axis([0, 4, 0, 3.5])

#Fill the region below the line
ax.fill_between(x0,x1, alpha=0.2)

#Plot the original data
plot_data(X, Y, ax)
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')
plt.show()
```
